# Steam Data Over Time

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
# Load and clean STEAM data
steam_survey = pd.read_csv('https://raw.githubusercontent.com/jdegene/steamHWsurvey/master/shs.csv')
steam_dates = steam_survey[pd.to_datetime(steam_survey['date']).dt.year >= 2015].reset_index().drop(labels='index',axis=1)
steam_gpu = steam_dates[steam_dates['category'] == 'Video Card Description']
steam_gpu = steam_gpu.dropna()
steam_gpu['date'] = pd.to_datetime(steam_gpu['date'])

steam_gpu.head()
```

```python
# AMD 
amd = steam_gpu[steam_gpu['name'].str.contains("AMD")]

plt.figure(figsize=(10, 6))
sns.lineplot(x='date', y='percentage', hue='name', data=amd)

plt.legend(loc='upper left', bbox_to_anchor=(1, 0.5))
plt.gca().xaxis.set_major_locator(mdates.YearLocator())

plt.title('AMD GPU Usage Over Time')
plt.xlabel('Date')
plt.ylabel('Percentage')

plt.show()
```

```python
# NVIDIA 
nvidia = steam_gpu[steam_gpu['name'].str.contains("NVIDIA")]
plt.figure(figsize=(10, 6))
sns.lineplot(x='date', y='percentage', hue='name', data=nvidia)

plt.legend(loc='upper left', bbox_to_anchor=(1, 0.5))
plt.gca().xaxis.set_major_locator(mdates.YearLocator())

plt.title('NVIDIA GPU Usage Over Time')
plt.xlabel('Date')
plt.ylabel('Percentage')

plt.show()
```

```python
# Intel 
intel = steam_gpu[steam_gpu['name'].str.contains("Intel")]
plt.figure(figsize=(10, 6))
sns.lineplot(x='date', y='percentage', hue='name', data=intel)

plt.legend(loc='upper left', bbox_to_anchor=(1, 0.5))
plt.gca().xaxis.set_major_locator(mdates.YearLocator())

plt.title('Intel GPU Usage Over Time')
plt.xlabel('Date')
plt.ylabel('Percentage')

plt.show()
```
